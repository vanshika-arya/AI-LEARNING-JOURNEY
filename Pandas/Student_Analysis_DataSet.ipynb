## Memory Optimization Using "category" Datatype in Pandas

```python
print(f"Original 'grade' dtype :{df['grade'].dtype}")
print("Before optimization memory usage")
print(df['grade'].memory_usage(deep=True)) #there are two types of category one is 'object' and 2nd is 'category'....obhject string me store karta hai isliye jyada space leta hai and category int me store karti hai isliye kam storage rahta hai
df['grade']=df['grade'].astype('category')
print(f"optimized 'grade' dtype: {df['grade'].dtype}")  #deep = true----> bina uske sirf ram check karega numpy....and deep = true me storage me bhi check karega!!!
print("Memory usage after optimization")
print(df['grade'].memory_usage(deep=True)) #agr cloumn me repeated values hai to category use karna cahiye
```

## 17. vector grade calculation from final_score

```python
#vector grade calculate from final_score
df['grade'] = 'F'
df['grade'][df['final_score'] >= 60] = 'D'
df['grade'][df['final_score'] >= 70] = 'C'
df['grade'][df['final_score'] >= 80] = 'B'
df['grade'][df['final_score'] >= 90] = 'A'
print(df[['final_score', 'grade']].head())
```

```python
def assign_grade(score):
    if score>=90: return 'A'
    elif score>=80: return 'B'
    elif score>=70: return 'C'
    elif score>=60: return 'D'
    else: return 'F'

df['calculated_grade']=df['final_score'].apply(assign_grade)
print(df[['final_score', 'calculated_grade']].tail())
```

## 18. 



```python
def assign_status(final_score):
    if final_score>=60:
        return 'Pass'
    else:
        return 'Fail'

df['calculate_status']= df['final_score'].apply(assign_status)
print(df[['final_score', 'calculate_status']].head())
```
