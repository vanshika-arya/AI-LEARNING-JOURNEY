```python
print(f"Original 'grade' dtype :{df['grade'].dtype}")
print("Before optimization memory usage")
print(df['grade'].memory_usage(deep=True)) #there are two types of category one is 'object' and 2nd is 'category'....obhject string me store karta hai isliye jyada space leta hai and category int me store karti hai isliye kam storage rahta hai
df['grade']=df['grade'].astype('category')
print(f"optimized 'grade' dtype: {df['grade'].dtype}")  #deep = true----> bina uske sirf ram check karega numpy....and deep = true me 
print("Memory usage after optimization")
print(df['grade'].memory_usage(deep=True))
```

```python
math_weight =0.4
science_weight=0.3
english_weight=0.3

df['coustom_weighted_score'] = (df['math_original']* math_weight + 
                               df['science_original']* science_weight +
                               df['english_original']* english_weight)
print("First 5 coustom weighted scores")
print(df[['math_original', 'science_original', 'english_original', 'coustom_weighted_score']].head())
```
